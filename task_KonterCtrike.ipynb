$$ \textbf{Критерий К-С} $$

```python
import pandas

# Посмотрим на гистораммочку

df1 = pandas.read_csv("experiments.csv")
df1['experiments'].plot(kind='bar')
```

```python
# Кинем взгляд на характеристечку распределеньичка

print(df1.experiments.describe())
```

```python
# Сейчас главное событие этого задания - график НОРМАЛЬНОГО распределения

df12 = pandas.DataFrame(data={
    'df1': df1['experiments']
})

df12.plot.kde()
```

```python
from scipy import stats

# Получим К-С критерии и сделаем выводы

d1 = df12['df1']
print(stats.kstest(d1, 'norm', (d1.mean(), d1.std()), N=5000))
```

Вывод: первое значение мало, второе великоо! Значит, распределение норм.
